Splitting the features into Categorical features and numerical features

```python
cat_column = df.columns[df.dtypes == 'object']
num_column = df.columns[df.dtypes != 'object']

print(cat_column)
print(num_column)
```

```python
df[cat_column].describe()
```

Finding the correlation between numerical features

```python
#plotting the correlation graph
plt.figure(figsize= (20,10))
sns.heatmap(data = df[num_column].corr(), annot = True)
```

```python
# plotting the numerical features
plt.figure(figsize= (5,5))
for i in df[num_column]:
    sns.histplot(data = df,x = i, kde = True)
    print('/n')
    plt.show()
```
